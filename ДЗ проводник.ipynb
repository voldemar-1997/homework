```python
import os
import shutil

def create_file(filename):
    try:
        with open(filename, 'w'):
            print(f"Файл '{filename}' успешно создан.")
    except IOError:
        print(f"Ошибка при создании файла '{filename}'.")

def delete_file(filename):
    try:
        os.remove(filename)
        print(f"Файл '{filename}' успешно удален.")
    except FileNotFoundError:
        print(f"Файл '{filename}' не найден.")
    except PermissionError:
        print(f"Нет разрешения на удаление файла '{filename}'.")

def move_file(source, destination):
    try:
        shutil.move(source, destination)
        print(f"Файл '{source}' успешно перемещен в '{destination}'.")
    except FileNotFoundError:
        print(f"Файл '{source}' не найден.")
    except PermissionError:
        print(f"Нет разрешения на перемещение файла '{source}'.")

def copy_file(source, destination):
    try:
        if os.path.isdir(source):
            shutil.copytree(source, destination)
            print(f"Папка '{source}' успешно скопирована в '{destination}'.")
        else:
            shutil.copy(source, destination)
            print(f"Файл '{source}' успешно скопирован в '{destination}'.")
    except FileNotFoundError:
        print(f"Файл или папка '{source}' не найдены.")
    except PermissionError:
        print(f"Нет разрешения на копирование '{source}'.")

def create_directory(directory):
    try:
        os.mkdir(directory)
        print(f"Папка '{directory}' успешно создана.")
    except FileExistsError:
        print(f"Папка '{directory}' уже существует.")
    except PermissionError:
        print(f"Нет разрешения на создание папки '{directory}'.")

def delete_directory(directory):
    try:
        os.rmdir(directory)
        print(f"Папка '{directory}' успешно удалена.")
    except FileNotFoundError:
        print(f"Папка '{directory}' не найдена.")
    except OSError:
        print(f"Папка '{directory}' не пуста или нет разрешения на удаление.")

def print_menu():
    print("1. Создать файл")
    print("2. Удалить файл")
    print("3. Переместить файл")
    print("4. Скопировать файл/папку")
    print("5. Создать папку")
    print("6. Удалить папку")
    print("0. Выход")

while True:
    print_menu()
    choice = input("Введите номер операции: ")

    if choice == "1":
        filename = input("Введите имя файла: ")
        create_file(filename)
    elif choice == "2":
        filename = input("Введите имя файла: ")
        delete_file(filename)
    elif choice == "3":
        source = input("Введите имя исходного файла: ")
        destination = input("Введите имя файла назначения: ")
        move_file(source, destination)
    elif choice == "4":
        source = input("Введите имя исходного файла: ")
        destination = input("Введите имя файла назначения: ")
        copy_file(source, destination)
    elif choice == "5":
        directory = input("Введите имя папки: ")
        create_directory(directory)
    elif choice == "6":
        directory = input("Введите имя папки: ")
        delete_directory(directory)
    elif choice == "0":
        break
    else:
        print("Неверный выбор. Попробуйте еще раз.")
```
